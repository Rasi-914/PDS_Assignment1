```python
data['Total_Score'] = data['math score'] + data['writing score'] + data['reading score']
```

```python
import matplotlib.pyplot as plt
```

```python
gender_counts = data['gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.savefig ('D:/R_Drive/UMKC/Subjects/Fall_Semester/Priciples_of_Datascience/Assignments/Assignment1/1.2/results/Gender_distribution.png')
plt.show()
```
